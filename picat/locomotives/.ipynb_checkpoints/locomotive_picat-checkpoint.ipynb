```python
%%picat
import input_sample.
import util.
main =>
   
   services(S),
   printf("\n %w \n", S),
   my_sort(S,L1),
   h_sort(S,L2), 
   L3 = sort_2(S),
   %write(L1), nl,
   printf("\n SORTED OK\n: %w  \n", L3), 
   print_list(L3), 
   printf("\n %w %w %w \n", len(S), len(L1), len(L3)) .

print_list([]) =>
    true.
print_list([ X | Xs]) =>
    printf("\n (%w),", X),
    %writeln(X),
    print_list(Xs).

%my_sort([], R) ?=> R  = [].
%my_sort([A], [R]), len([A]) == 1  ?=>  R = A.
my_sort([], A) ?=> A = [].
my_sort([A], AL) ?=> AL = [A].

my_sort([A,B|L], Menor_L1) ?=>
   %[Menor|L1] = Menor_L1, % fixed by hakank
    (_,_,Xa,_,_) = A,
    (_,_,Xb,_,_) = B,
    Xa @< Xb, % hakank: Using @< instead of <
    Menor = A,
    Menor_L1 = [Menor|L1],
    my_sort([B|L], L1).

my_sort([A,B|L], Menor_L1) =>
    %[Menor|L1] = Menor_L1,  % fixed by hakank
    (_,_,Xa,_,_) = A,
    (_,_,Xb,_,_) = B,
    Xa @>= Xb,
    Menor = B,
    Menor_L1 = [Menor|L1],
    my_sort([A|L], L1).

%%% by Hakan
h_sort([], []).
h_sort([A], [A]).
h_sort([A,B|L], [Menor|L1]) :-
    (_,_,Xa,_,_) = A,
    (_,_,Xb,_,_) = B,
    ( Xa @< Xb ->
       Menor = A,
       h_sort([B|L], L1)
    ;
      Menor = B,    
      h_sort([A|L], L1)
    ).


n_3_tupla((_, _, X, _, _)) = X.

menor([]) = R, len([]) == 0 => R = 0.
menor([A]) = R, len([A]) == 1 => R = A.              
menor([A, B | L]) = R, n_3_tupla(A) @< n_3_tupla(B) => R = menor([A | L]).    
menor([A, B | L]) = R, n_3_tupla(B) @=< n_3_tupla(A) => R = menor([B | L]).

sort_2([]) = R => R = [].
sort_2([A]) = R => R = [A].
sort_2(L) = R => 
    M = menor(L),
    L_2 = delete(L, M),
    R = [ M | sort_2(L_2) ].
```
